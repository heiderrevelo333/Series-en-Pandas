```python
#Aplicación de funciones: apply y map
```

```python
serie = pd.Series([2, 4, 6, 8])

cuadrado = serie.apply(lambda x: x**2)
print(cuadrado)
```

```python
nombres = pd.Series(["juan", "ana", "luis"])

mayusculas = nombres.map(str.upper)
print(mayusculas)
```

```python
numeros = pd.Series([4, 9, 16, 25])

raiz = numeros.apply(np.sqrt)
print(raiz)
```

```python
def clasificar(valor):
    if valor >= 50:
        return "Alto"
    else:
        return "Bajo"

datos = pd.Series([20, 45, 60, 80])

resultado = datos.apply(clasificar)
print(resultado)
```

```python
colores = pd.Series(["rojo", "azul", "verde"])

diccionario = {
    "rojo": "R",
    "azul": "A",
    "verde": "V"
}

reemplazo = colores.map(diccionario)
print(reemplazo)
```

### Uso de Series en Pandas
#### Eficiencia y Comprensión de los Datos
### Eficiencia en el manejo de datos
Estructura optimizada: Una Series es un array unidimensional que almacena datos de un
solo tipo (por ejemplo, enteros, flotantes o cadenas). Esto permite realizar operaciones
vectorizadas, las cuales son mucho más rápidas que los bucles tradicionales en Python.
Operaciones rápidas: Pandas utiliza bibliotecas como NumPy para ejecutar cálculos
optimizados en memoria. Esto acelera operaciones matemáticas, estadísticas y de
agregación, incluso cuando se trabaja con grandes volúmenes de datos.
Integración con DataFrames: Las Series son los bloques básicos de los DataFrame (tablas de
datos). Cada columna de un DataFrame es una Series, lo que permite manipular columnas
individuales de forma eficiente.
### Comprensión y claridad
Etiquetas descriptivas: Cada elemento en una Series tiene un índice o etiqueta. Esto permite
acceder a los datos de forma clara e intuitiva, por ejemplo: serie['nombre_etiqueta'].
Visualización sencilla: Las Series son fáciles de imprimir y visualizar, ya que muestran los
valores junto con sus índices, lo que ayuda a entender rápidamente la estructura de los
datos.
Funciones integradas: Pandas incluye métodos como .describe(), .mean(), .max() o .plot()
que permiten obtener estadísticas y gráficos con una sola línea de código, facilitando el
análisis exploratorio.
### Conclusión
Las Series en Pandas son esenciales en el análisis de datos porque combinan rendimiento
(gracias a su estructura optimizada y al uso de NumPy) y claridad (mediante etiquetas
descriptivas y funciones integradas). Esto las convierte en una herramienta clave para
cualquier proyecto de análisis de datos, desde la limpieza hasta la visualización.
