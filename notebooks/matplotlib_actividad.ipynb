## Sección A – Estructura de una Gráfica

```python
# A.1 – Crear una Figura con plt.figure()
# La Figura es el contenedor principal de todos los elementos gráficos.

fig = plt.figure(figsize=(8, 5))

# Agregamos un subplot manualmente (axes)
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color='steelblue', label='Seno')

# Elementos de estructura completa
ax.set_title('Figura creada con plt.figure()', fontsize=14, fontweight='bold')
ax.set_xlabel('Eje X – tiempo')
ax.set_ylabel('Eje Y – amplitud')
ax.legend(loc='upper right')
ax.annotate('Máximo', xy=(np.pi/2, 1), xytext=(2.5, 1.2),
            arrowprops=dict(arrowstyle='->', color='red'), color='red')
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('grafica_A1_figure.png', dpi=150, bbox_inches='tight')
plt.show()
print('Gráfica guardada como grafica_A1_figure.png')
```

```python
# A.2 – Crear Figura y Ejes con plt.subplots()
# plt.subplots() crea la figura y uno o más objetos Axes en una sola llamada.

fig, ax = plt.subplots(figsize=(8, 5))

x = np.linspace(0, 2 * np.pi, 200)
ax.plot(x, np.cos(x), color='darkorange', linewidth=2, label='Coseno')

ax.set_title('Figura creada con plt.subplots()', fontsize=14, fontweight='bold')
ax.set_xlabel('Radianes')
ax.set_ylabel('Valor')
ax.legend()
ax.grid(True, alpha=0.4)

plt.tight_layout()
plt.savefig('grafica_A2_subplots.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# A.3 – Crear ejes con add_subplot() en cuadrícula múltiple
# add_subplot(filas, columnas, índice) permite posicionar ejes con precisión.

fig = plt.figure(figsize=(10, 4))

ax1 = fig.add_subplot(1, 2, 1)  # Fila 1, Columna 1
ax2 = fig.add_subplot(1, 2, 2)  # Fila 1, Columna 2

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), 'b-')
ax1.set_title('Seno – add_subplot(1,2,1)')
ax1.set_xlabel('x'); ax1.set_ylabel('sin(x)')

ax2.plot(x, x**2, 'r-')
ax2.set_title('Cuadrática – add_subplot(1,2,2)')
ax2.set_xlabel('x'); ax2.set_ylabel('x²')

plt.tight_layout()
plt.savefig('grafica_A3_add_subplot.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# A.4 – Título, Etiquetas, Leyendas y Anotaciones completos

fig, ax = plt.subplots(figsize=(9, 5))

x = np.linspace(0, 4 * np.pi, 300)
y1 = np.sin(x)
y2 = np.sin(x) * np.exp(-0.15 * x)

ax.plot(x, y1, label='Seno puro', color='royalblue', linestyle='-')
ax.plot(x, y2, label='Seno amortiguado', color='tomato', linestyle='--')

# Título
ax.set_title('Estructura completa de una gráfica en Matplotlib', fontsize=15, fontweight='bold', pad=15)

# Etiquetas de ejes
ax.set_xlabel('Tiempo (radianes)', fontsize=12)
ax.set_ylabel('Amplitud', fontsize=12)

# Leyenda
ax.legend(loc='upper right', fontsize=11, framealpha=0.9)

# Anotación con flecha
ax.annotate('Primer máximo\nde seno puro',
            xy=(np.pi/2, 1.0), xytext=(1.5, 1.3),
            arrowprops=dict(arrowstyle='->', color='navy', lw=1.5),
            fontsize=10, color='navy')

ax.grid(True, linestyle=':', alpha=0.6)
ax.set_ylim(-1.5, 1.6)

plt.tight_layout()
plt.savefig('grafica_A4_estructura_completa.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Sección C – Gráficas Univariantes

```python
# C.1 – Histograma básico
# Muestra la distribución de frecuencias de una variable numérica continua.

np.random.seed(42)
datos = np.random.randn(500)  # 500 valores con distribución normal

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(datos, bins=25, color='steelblue', edgecolor='white')
ax.set_title('C.1 – Histograma básico')
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C1_histograma_basico.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.2 – Histograma acumulativo
# Muestra cuántas observaciones acumuladas hay hasta cada valor.

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(datos, bins=25, cumulative=True, color='mediumseagreen', edgecolor='white')
ax.set_title('C.2 – Histograma acumulativo')
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia acumulada')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C2_histograma_acumulativo.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.3 – Histograma normalizado (density=True)
# El área total bajo el histograma equivale a 1 (distribución de probabilidad).

fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(datos, bins=25, density=True, color='mediumpurple', edgecolor='white', alpha=0.8)

# Curva de densidad superpuesta
xn = np.linspace(datos.min(), datos.max(), 200)
mu = datos.mean()
sigma = datos.std()
pdf_normal = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((xn - mu) / sigma)**2)
ax.plot(xn, pdf_normal, 'r-', linewidth=2, label='PDF normal')

ax.set_title('C.3 – Histograma normalizado (density=True)')
ax.set_xlabel('Valor')
ax.set_ylabel('Densidad de probabilidad')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C3_histograma_normalizado.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.4 – Boxplot básico
# Muestra mediana, cuartiles y valores atípicos de una distribución.

np.random.seed(10)
grupos = [np.random.normal(0, 1, 100),
          np.random.normal(2, 1.5, 100),
          np.random.normal(-1, 0.8, 100)]

fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot(grupos, tick_labels=['Grupo A', 'Grupo B', 'Grupo C'])
ax.set_title('C.4 – Boxplot básico')
ax.set_ylabel('Valor')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C4_boxplot_basico.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.5 – Boxplot personalizado con colores

grupos = [np.random.normal(0, 1, 100), np.random.normal(2, 1.5, 100), np.random.normal(-1, 0.8, 100)]

colores = ['red', 'green', 'blue']

fig, ax = plt.subplots(figsize=(7, 5))

bp = ax.boxplot(grupos, patch_artist=True,  # patch_artist=True para poder colorear
                tick_labels=['Grupo A', 'Grupo B', 'Grupo C'],
                medianprops=dict(color='red', linewidth=2))

for patch, color in zip(bp['boxes'], colores):
    patch.set_facecolor(color)
    patch.set_alpha(0.8)

ax.set_title('C.5 – Boxplot personalizado con colores')
ax.set_ylabel('Valor')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C5_boxplot_color.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.6 – Violin Plot básico
# Combina boxplot con estimación de densidad de kernel (KDE).

grupos = [np.random.normal(0, 1, 100), np.random.normal(2, 1.5, 100), np.random.normal(-1, 0.8, 100)]

fig, ax = plt.subplots(figsize=(7, 5))
ax.violinplot(grupos, positions=[1, 2, 3], showmedians=True)
ax.set_title('C.6 – Violin Plot básico')
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Grupo A', 'Grupo B', 'Grupo C'])
ax.set_ylabel('Valor')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C6_violin_basico.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.7 – Violin Plot personalizado

grupos = [np.random.normal(0, 1, 100), np.random.normal(2, 1.5, 100), np.random.normal(-1, 0.8, 100)]

fig, ax = plt.subplots(figsize=(7, 5))
parts = ax.violinplot(grupos, positions=[1, 2, 3],
                      showmeans=True, showmedians=True, showextrema=False)

colores_violin = ['#42A5F5', '#66BB6A', '#FFA726']
for pc, color in zip(parts['bodies'], colores_violin):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

parts['cmeans'].set_color('navy')
parts['cmedians'].set_color('red')

ax.set_title('C.7 – Violin Plot personalizado')
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Grupo A', 'Grupo B', 'Grupo C'])
ax.set_ylabel('Valor')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_C7_violin_personalizado.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.8 – Gráfica de línea simple (función seno)

x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x), color='dodgerblue', linewidth=2)
ax.set_title('C.8 – Gráfica de línea simple (seno)')
ax.set_xlabel('x (radianes)')
ax.set_ylabel('sin(x)')
ax.axhline(0, color='gray', linewidth=0.8, linestyle='--')  # Línea de referencia en y=0
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafica_C8_linea_seno.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.9 – Gráficas de línea personalizadas (colores, estilos, marcadores)

x = np.linspace(0, 10, 15)  # Menos puntos para que los marcadores sean visibles

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(x, np.sin(x),   color='royalblue',  linestyle='-',  marker='o', markersize=6, label='seno – sólido + círculo')
ax.plot(x, np.cos(x),   color='tomato',     linestyle='--', marker='s', markersize=6, label='coseno – guion + cuadrado')
ax.plot(x, np.tan(x)/5, color='seagreen',   linestyle=':',  marker='^', markersize=6, label='tangente/5 – punteado + triángulo')

ax.set_title('C.9 – Líneas con colores, estilos y marcadores')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_ylim(-1.5, 1.5)
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafica_C9_lineas_personalizadas.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# C.10 – Gráfica de línea comparativa (seno vs coseno)

x = np.linspace(0, 2 * np.pi, 200)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x, np.sin(x), color='royalblue',  linewidth=2.5, label='sin(x)')
ax.plot(x, np.cos(x), color='darkorange', linewidth=2.5, linestyle='--', label='cos(x)')

ax.fill_between(x, np.sin(x), np.cos(x), alpha=0.15, color='purple', label='Diferencia')

ax.set_title('C.10 – Gráfica comparativa: seno vs coseno')
ax.set_xlabel('x (radianes)')
ax.set_ylabel('Amplitud')
ax.legend()
ax.axhline(0, color='gray', linewidth=0.8, linestyle=':')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafica_C10_comparativa_seno_coseno.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# D.4 – Scatter con línea de regresión lineal

fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(x_sc, y_sc, color='cornflowerblue', alpha=0.6, label='Datos')

# Regresión lineal con numpy
m, b = np.polyfit(x_sc, y_sc, 1)
x_line = np.linspace(x_sc.min(), x_sc.max(), 100)
ax.plot(x_line, m * x_line + b, color='red', linewidth=2,
        label=f'Regresión: y = {m:.2f}x + {b:.2f}')

ax.set_title('D.4 – Scatter con línea de regresión lineal')
ax.set_xlabel('X'); ax.set_ylabel('Y')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafica_D4_scatter_regresion.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# D.6 – Línea con regresión cuadrática (polinomio grado 2)

np.random.seed(5)
x6 = np.linspace(-3, 3, 80)
y6 = 2 * x6**2 - 3 * x6 + 1 + np.random.randn(80) * 1.5

coefs = np.polyfit(x6, y6, 2)  # Ajuste cuadrático
p = np.poly1d(coefs)
x_curva = np.linspace(-3, 3, 200)

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(x6, y6, color='steelblue', alpha=0.6, s=40, label='Datos observados')
ax.plot(x_curva, p(x_curva), color='red', linewidth=2.5,
        label=f'Regresión cuadrática\ny = {coefs[0]:.2f}x² + {coefs[1]:.2f}x + {coefs[2]:.2f}')
ax.set_title('D.6 – Scatter con regresión cuadrática')
ax.set_xlabel('X'); ax.set_ylabel('Y')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafica_D6_regresion_cuadratica.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# D.9 – Heatmap de matriz de correlación

np.random.seed(42)
df = pd.DataFrame({
    'Precio':      np.random.randn(100),
    'Ventas':      np.random.randn(100) * 2,
    'Publicidad':  np.random.randn(100) * 0.5,
    'Clientes':    np.random.randn(100) * 1.5,
    'Satisfacción':np.random.randn(100)
})
# Crear correlaciones artificiales
df['Ventas'] = df['Precio'] * 0.8 + np.random.randn(100) * 0.5
df['Clientes'] = df['Ventas'] * 0.6 + np.random.randn(100) * 0.8

corr = df.corr()
variables = corr.columns.tolist()

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr.values, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax, label='Coeficiente de correlación')

ax.set_xticks(range(len(variables))); ax.set_xticklabels(variables, rotation=30, ha='right')
ax.set_yticks(range(len(variables))); ax.set_yticklabels(variables)
ax.set_title('D.9 – Heatmap de matriz de correlación')
plt.tight_layout()
plt.savefig('grafica_D9_correlacion.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# D.10 – Heatmap con anotaciones numéricas

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr.values, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax, label='Correlación')

# Anotar cada celda con el valor numérico
for i in range(len(variables)):
    for j in range(len(variables)):
        val = corr.values[i, j]
        color_texto = 'white' if abs(val) > 0.5 else 'black'
        ax.text(j, i, f'{val:.2f}', ha='center', va='center', fontsize=9, color=color_texto)

ax.set_xticks(range(len(variables))); ax.set_xticklabels(variables, rotation=30, ha='right')
ax.set_yticks(range(len(variables))); ax.set_yticklabels(variables)
ax.set_title('D.10 – Heatmap con anotaciones numéricas')
plt.tight_layout()
plt.savefig('grafica_D10_heatmap_anotado.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Sección E – Gráficas Categóricas

```python
# Datos compartidos para Sección E
categorias = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
trimestre1 = np.array([120, 95, 140, 80])
trimestre2 = np.array([135, 110, 130, 100])
trimestre3 = np.array([150, 125, 115, 120])
x_pos = np.arange(len(categorias))
```

```python
# E.1 – Barras apiladas
# Muestra la composición total por categoría, apilando segmentos.

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(categorias, trimestre1, label='T1', color='#42A5F5')
ax.bar(categorias, trimestre2, bottom=trimestre1, label='T2', color='#66BB6A')
ax.bar(categorias, trimestre3, bottom=trimestre1 + trimestre2, label='T3', color='#FFA726')

ax.set_title('E.1 – Barras apiladas por trimestre')
ax.set_xlabel('Producto')
ax.set_ylabel('Ventas totales')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_E1_barras_apiladas.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# E.2 – Barras agrupadas
# Permite comparar valores de distintas series lado a lado por categoría.

ancho = 0.25

fig, ax = plt.subplots(figsize=(9, 5))
ax.bar(x_pos - ancho, trimestre1, width=ancho, label='T1', color='#42A5F5')
ax.bar(x_pos,         trimestre2, width=ancho, label='T2', color='#66BB6A')
ax.bar(x_pos + ancho, trimestre3, width=ancho, label='T3', color='#FFA726')

ax.set_title('E.2 – Barras agrupadas por trimestre')
ax.set_xlabel('Producto')
ax.set_ylabel('Ventas')
ax.set_xticks(x_pos)
ax.set_xticklabels(categorias)
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_E2_barras_agrupadas.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# E.3 – Barras apiladas personalizadas (colores, hatch/tramas)

colores_apil = ['#1565C0', '#2E7D32', '#E65100']
tramas = ['/', 'x', 'o']

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(categorias, trimestre1, label='T1', color=colores_apil[0], hatch=tramas[0], edgecolor='white')
ax.bar(categorias, trimestre2, bottom=trimestre1,
       label='T2', color=colores_apil[1], hatch=tramas[1], edgecolor='white')
ax.bar(categorias, trimestre3, bottom=trimestre1 + trimestre2,
       label='T3', color=colores_apil[2], hatch=tramas[2], edgecolor='white')

ax.set_title('E.3 – Barras apiladas personalizadas (colores + hatch)')
ax.set_xlabel('Producto')
ax.set_ylabel('Ventas')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_E3_barras_apiladas_custom.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# E.4 – Barras con etiquetas numéricas sobre cada segmento

fig, ax = plt.subplots(figsize=(8, 5))

bar1 = ax.bar(categorias, trimestre1, label='T1', color='#42A5F5')
bar2 = ax.bar(categorias, trimestre2, bottom=trimestre1, label='T2', color='#66BB6A')
bar3 = ax.bar(categorias, trimestre3, bottom=trimestre1 + trimestre2, label='T3', color='#FFA726')

# Función para añadir etiquetas en el centro de cada segmento
def etiquetar_segmentos(bars, base_arr):
    for bar, base in zip(bars, base_arr):
        h = bar.get_height()
        if h > 0:
            ax.text(bar.get_x() + bar.get_width() / 2,
                    base + h / 2, str(int(h)),
                    ha='center', va='center', fontsize=9, color='white', fontweight='bold')

etiquetar_segmentos(bar1, np.zeros(len(categorias)))
etiquetar_segmentos(bar2, trimestre1)
etiquetar_segmentos(bar3, trimestre1 + trimestre2)

ax.set_title('E.4 – Barras apiladas con etiquetas numéricas')
ax.set_xlabel('Producto')
ax.set_ylabel('Ventas')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('grafica_E4_barras_etiquetas.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 1. ¿Cuál gráfica fue más útil para interpretar datos y por qué?

El **heatmap de matriz de correlación con anotaciones (D.10)** resultó especialmente útil porque permite identificar de un vistazo las relaciones entre múltiples variables numéricas simultáneamente. El uso de color (rojo para correlación positiva fuerte, azul para negativa) y los valores numéricos superpuestos elimina la ambigüedad y facilita la toma de decisiones analíticas sin necesidad de calcular cada par manualmente.

### 2. ¿Qué diferencias notaste entre usar pyplot y la API OO?

| Aspecto | Interfaz pyplot (`plt.*`) | Interfaz OO (`fig, ax`) |
|---|---|---|
| **Simplicidad** | Alta – código más corto | Moderada – más verbosa |
| **Control** | Limitado – estado global | Total – acceso directo a cada objeto |
| **Reutilización** | Difícil en funciones/clases | Ideal para encapsulamiento |
| **Subgráficas** | Complicado con varias | Natural con `axs[i]` |
| **Recomendado para** | Exploración rápida | Producción y proyectos |

La API OO es claramente superior cuando se trabaja con múltiples ejes, se reutilizan componentes o se genera código modular. pyplot es suficiente para visualizaciones exploratorias de una sola gráfica.

### 3. ¿Qué tipo de gráfica consideras más adecuada para tu proyecto personal?

Para un proyecto de análisis de datos de ventas, consideraría combinar:
- **Líneas comparativas (C.10 / D.5)**: para seguimiento de tendencias temporales entre productos o periodos.
- **Barras agrupadas (E.2)**: para comparar rendimiento por categoría en distintos trimestres.
- **Heatmap de correlación (D.9/D.10)**: para identificar variables relacionadas (precio, publicidad, ventas) antes de modelar.

Esta combinación ofrece perspectiva temporal, comparación categórica y análisis de relaciones en un solo dashboard analítico.
